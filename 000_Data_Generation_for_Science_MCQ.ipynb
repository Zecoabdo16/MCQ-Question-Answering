```python
import re
```

```python

def remove_non_english_characters(text):
    # Remove non-English characters
    english_text = re.sub(r'[^\x00-\x7F]+', '', text)

    return english_text

# Example usage
# input_text = "cnjλnjtc−Ndisplaystyle tjlambda  over njNnc over njlambda  over njtcN\nwhile njncnairdisplaystyle njncnair\nARROWs"
# english_text = remove_non_english_characters(input_text)

# print("English Text:")
# print(english_text)
```

```python
text_list = remove_non_english_characters(text_list)
```

```python
def remove_newlines(text):
    # Remove newline characters
    text = text.replace('\n', '')

    return text
```

```python
clean_text = remove_newlines(text_list)
```

```python
def split_text_by_words(long_text, max_words):
    # Split the long text into a list of shorter texts
    # with a maximum number of words of max_words
    words = long_text.split()
    shorter_texts = []
    current_text = ''

    for word in words:
        if len(current_text.split()) + 1 <= max_words:
            current_text += word + ' '
        else:
            shorter_texts.append(current_text.strip())
            current_text = word + ' '

    if current_text:
        shorter_texts.append(current_text.strip())

    return shorter_texts
```

```python
input_text = clean_text
max_words = 500

shorter_texts = split_text_by_words(input_text, max_words)
```

```python
prompt = []
context = []
A = []
B = []
C = []
D = []
E = []
answer = []

for load in loads[1000:2000]:
  output = qg.predict_mcq(load)
  try:
    for i in range( len( output['questions'] ) ):
      try:
        prompt.append(output['questions'][i]['question_statement'])
      except:
        prompt.append(np.nan)

      try:
        context.append(output['questions'][i]['context'])
      except:
        context.append(np.nan)

      try:
        A.append(output['questions'][i]['options'][0])
      except:
        A.append(np.nan)

      try:
        B.append(output['questions'][i]['options'][1])
      except:
        B.append(np.nan)

      try:
        C.append(output['questions'][i]['options'][2])
      except:
        C.append(np.nan)

      try:
        D.append(output['questions'][i]['answer'])
      except:
        D.append(np.nan)

      try:
        E.append(output['questions'][i]['extra_options'][0])
      except:
        E.append(np.nan)
  except:
    continue
```

```python
len(A)
```

```python
data = {
    'prompt': prompt,
    'context' : context,
    'A': A,
    'B': B,
    'C' : C,
    'D' : D,
    'E' : E
    # 'answer' : answer
}

df = pd.DataFrame(data)
```
